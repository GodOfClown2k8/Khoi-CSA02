```python
import pandas as pd 
df = pd.read_excel('provinces.xlsx') 
df
```

## Cau 1 


```python
population_in_persons = df['Population'] * 1000
df['Density'] = population_in_persons / df['Area']
df
```

```python
highest_density = df.loc[df['Density'].idxmax()]
highest_density
```

```python
lowest_density = df.loc[df['Density'].idxmin()]
lowest_density
```

## Cau 2

```python
coastlines_df = pd.read_excel('coastlines.xlsx')

num_coastal_provinces = len(coastlines_df)
print(f"Số tỉnh thành giáp biển ở Việt Nam: {num_coastal_provinces}")

for index, row in coastlines_df.iterrows():
    province_name = row['Name'] 
    coastline_length = row['Coastline']  
    df.loc[df['Name'] == province_name, 'Coastline'] = coastline_length


coastlines_df.to_excel('provinces_updated.xlsx', index=False)
coastlines_df
```
