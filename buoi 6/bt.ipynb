```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('provinces_updated.xls')
df
```

## Cau 1

```python
df = df.sort_values('Population')
df
```

```python
province_indices = []
for i in range(len(df)):
    province_indices.append(f"{i + 1}")

plt.figure(figsize=(12, 6))
plt.bar(province_indices, df['Population'], color='skyblue')  # Biểu đồ ngang
plt.xlabel('Dân số' ,fontsize = 12)
plt.title('Dân số các tỉnh thành Việt Nam')
#plt.xticks(rotation=180)  # Xoay nhãn trục x nếu cần
plt.show()
```

## Cau 2

```python
region_df = pd.read_excel('region_areas.xls')
region_df

```

```python
regions = region_df['Region']
areas = region_df['Area Proportion (%)']

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 8))
plt.pie(areas, labels=regions, autopct='%1.1f%%', startangle=140)
plt.title('Tỉ lệ diện tích của các vùng địa lý')
plt.axis('equal')  # Để biểu đồ tròn có hình tròn
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
two_df = pd.read_csv('cyclones.csv')

# Kiểm tra dữ liệu
two_df.head()
```

```python


years = two_df['date']
events = two_df['cyclone_events']

plt.figure(figsize=(10,6))
plt.plot(years, events, marker='o', color='orange')
plt.title('Cyclone Events in Vietnam', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Events', fontsize=12)
plt.grid(True)
plt.show()
```
